```python
from haversine import haversine, Unit
import haversine as hs
```

```python
platoon_alldist = []
platoon_all_speed = []

platoon_all_lat_dist = []
platoon_all_lon_dist = []

platoon_all_lat_speed = []
platoon_all_lon_speed = []



for i in range(len(ps)):
    platoon_lat_dist = []
    platoon_lon_dist = []
    
    platoon_dist = []
    speed = []

    
#ps - jc
    ps_ = ps.iloc[i]
    
    df_array1 = (ps_.Latitude, ps_.Longtitude)
    
    delta_t = jc['Time[s]'] - ps_['Time[s]']
    delta_pos = delta_t[delta_t > 0]
    min_delta = min(delta_pos)
    min_delta_index = delta_t[delta_t == min_delta].index
    
    jc_ = jc.loc[min_delta_index]
        
    df_array2 = (jc_.Latitude.iloc[0], jc_.Longtitude.iloc[0])
    
    platoon_dist.append(hs.haversine(df_array1,df_array2,unit=Unit.METERS))

    platoon_lat_dist.append(list(jc_["Latitude"])[0])
    platoon_lon_dist.append(list(jc_["Longtitude"])[0])
    
    speed = []
    speed.extend(list(jc_['Speed[m/s]']))  

#jc - rw
    
    delta_t = rw['Time[s]'] - jc_['Time[s]'].values
    delta_pos = delta_t[delta_t > 0]
    min_delta = min(delta_pos)
    min_delta_index = delta_t[delta_t == min_delta].index
    rw_ = rw.loc[min_delta_index]
        
    df_array3 = (rw_.Latitude.iloc[0], rw_.Longtitude.iloc[0])
    
    platoon_dist.append(hs.haversine(df_array2,df_array3,unit=Unit.METERS))   
    platoon_lat_dist.append(list(rw_["Latitude"])[0])
    platoon_lon_dist.append(list(rw_["Longtitude"])[0])

    speed.extend(list(rw_['Speed[m/s]']))   

 #rw-dw   
    
    delta_t = dw['Time[s]'] - rw_['Time[s]'].values
    delta_pos = delta_t[delta_t > 0]
    min_delta = min(delta_pos)
    min_delta_index = delta_t[delta_t == min_delta].index
    dw_ = dw.loc[min_delta_index]
        
    df_array4 = (dw_.Latitude.iloc[0], dw_.Longtitude.iloc[0])
    
    platoon_dist.append(hs.haversine(df_array3,df_array4,unit=Unit.METERS))   
    platoon_lat_dist.append(list(dw_["Latitude"])[0])
    platoon_lon_dist.append(list(dw_["Longtitude"])[0])
    
    speed.extend(list(dw_['Speed[m/s]'])) 

#dw - aa  
    
    delta_t = aa['Time[s]'] - dw_['Time[s]'].values
    delta_pos = delta_t[delta_t > 0]
    min_delta = min(delta_pos)
    min_delta_index = delta_t[delta_t == min_delta].index
    aa_ = aa.loc[min_delta_index]
        
    df_array5 = (aa_.Latitude.iloc[0], aa_.Longtitude.iloc[0])
    
    platoon_dist.append(hs.haversine(df_array4,df_array5,unit=Unit.METERS))   
    platoon_lat_dist.append(list(aa_["Latitude"])[0])
    platoon_lon_dist.append(list(aa_["Longtitude"])[0])
    speed.extend(list(aa_['Speed[m/s]'])) 
    
 #aa - ak  
    delta_t = ak['Time[s]'] - aa_['Time[s]'].values
    delta_pos = delta_t[delta_t > 0]
    
    min_delta = min(delta_pos)
    min_delta_index = delta_t[delta_t == min_delta].index
    ak_ = ak.loc[min_delta_index]
        
    df_array6 = (ak_.Latitude.iloc[0], ak_.Longtitude.iloc[0])
    
    platoon_dist.append(hs.haversine(df_array5,df_array6,unit=Unit.METERS))   
    platoon_lat_dist.append(list(ak_["Latitude"])[0])
    platoon_lon_dist.append(list(ak_["Longtitude"])[0])
    speed.extend(list(ak_['Speed[m/s]'])) 

    
    platoon_all_lat_dist.append(platoon_lat_dist)
    platoon_all_lon_dist.append(platoon_lon_dist)
    platoon_alldist.append(platoon_dist)
    platoon_all_speed.append(speed)
```
